```python
import pandas as pd 

# 구글 드라이브 경로에서 read하기 위한 경로로 수정합니다.
def chage_google_drive(drive_path):
  result_path = url2='https://drive.google.com/uc?id=' + drive_path.split('/')[-2]
  return result_path

# data의 공유 path를 불러옵니다.
KOSPI_fundamental_path = 'https://drive.google.com/file/d/1-jbeRXYkx-S-AS1-l49gskXLvnh6sZyn/view?usp=sharing'

# 데이터를 read합니다.
KOSPI_fundamental = pd.read_csv(chage_google_drive(KOSPI_fundamental_path))
KOSPI_fundamental.head()
```

```python
import pandas as pd
from datetime import datetime

datelist = pd.date_range('2016-01-01', '2021-10-20', freq ='BY')
```

- BPS[주당순자산가치 (Book-value per share)]: 주당 순자산가치는 기업의 순자산을 발행 주식수로 나눈 것이
- PER : 주가 / 주당 순이익  (버는 것에 비해 가격이 어떤가?)
- PBR : 주가 / BPS -> 주가 순자산 비율 -> 자산에 비해 주가가 얼마나 비싼가?
    - PBR > 1 : 고평가
    - PBR < 1 : 저평가
- EPS :1주당 이익
- DIV 배당수익률 (Divedend yield ratio) : 주당 배당금 / 현재주가
- DPS (Divedend per share) : 1 주당 배당금

```python
# 01. 최근 5년간 배당을 한 기업
```

```python
corp_df = KOSPI_fundamental.pivot_table(index = 'Date', columns = '티커', values = 'DPS')
```

```python
div_count = corp_df.loc['2016':'2020',:].astype(bool).sum(axis=0).sort_values() # 0제외한 배당 횟수 조회
```

```python
div_count = div_count.to_frame().rename(columns = {0 : '배당횟수'})
```

```python
div_count = div_count[div_count['배당횟수'] == 5] # 5년 연속 배당한 기업
```

```python
div_corp_list = div_count.index
```

```python
# 02. 배당액을 꾸준하게 상승한 기업
```

```python
div_corp_2 = corp_df.loc['2016':'2020',div_corp_list]
div_corp_2
```

```python
div_ratio = div_corp_2.pct_change() # 전해 대비 배당 증가율
div_ratio
```

```python
# 매년 배당액이 상승한 기업
div_ratio = div_ratio[div_ratio>0].loc[2017:].dropna(axis = 1)
div_ratio
```

```python
corp_list_2 = div_ratio.columns.to_list()
corp_list_2
```

```python
# 03. 주당 순이익 증가율이 큰 순서대로 정렬해보자.
```

```python
corp = KOSPI_fundamental[KOSPI_fundamental['티커'].isin(corp_list_2)]
EPS = corp.pivot_table(index = 'Date', columns = '티커', values = 'EPS')
EPS
```

```python
EPS_pct_change = EPS.pct_change()
EPS_pct_change = EPS_pct_change.loc[2017:,].dropna(axis = 1)
EPS_pct_change[EPS_pct_change >0].dropna(axis = 1)
```
